```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams["figure.figsize"] = (10,5)
```

```python
x = [2007,2008,2009,2010,2011,2012,2013,2014]

Hispanic = [1418,1445,1349,1316,1348,1263,1261,1230]
White = [7050,6836,6297,5351,5016,4719,4535,4507]
Black = [2722,2725,2535,2282,2243,2269,2365,2194]
```

```python
plt.plot( x, Hispanic, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4, label="Hispanic")
plt.plot( x, White, marker='', color='olive', linewidth=4, label="White")
plt.plot( x, Black, marker='', color='Black', linewidth=4, linestyle='dashed', label="Black-Non-Hispanic")
plt.legend()
plt.title("Different Ethnicities in NY Who Suffered with Heart Disease over Time ")
plt.xlabel("Years")
plt.ylabel("Number of Deaths")
```
